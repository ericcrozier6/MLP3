```python
import pydub 
import numpy as np
import ffmpeg
from mutagen.mp3 import MP3
import os
import scipy.sparse
```

```python
#function to turn mp3 into npz file
#Used code from here:
#https://stackoverflow.com/questions/53633177/how-to-read-a-mp3-audio-file-into-a-numpy-array-save-a-numpy-array-to-mp3

def read(f, normalized=False):
    """MP3 to numpy array"""
    a = pydub.AudioSegment.from_mp3(f)
    y = np.array(a.get_array_of_samples())
    if a.channels == 2:
        y = y.reshape((-1, 2))
    if normalized:
        return a.frame_rate, np.float32(y) / 2**15
    else:
        return a.frame_rate, y
```

```python
# Get all names of files in directory
path = "C://Users//ericc//Desktop//MLP3//project3//test"
files = os.listdir(path)
 
# prints all files
#print(files)
```

```python
#finds the shortest audio length of all the mp3 files
#so we know if we need to cut any down so they are all uniform
lowestTime = 100.0

for i in files:
    audio = MP3("project3//test//" + i)
    if(float(audio.info.length) < float(lowestTime)):
        lowestTime = float(audio.info.length)

print(lowestTime)        
```

```python
from pydub import AudioSegment

#remove amount of audio from mp3 to make all files in directory equal length
#based on the shortest audio

#loop through all files in directory
for i in files:
    #get the mp3 file
    audio = AudioSegment.from_mp3("project3//test//" + i)
    startTime = 0
    endTime = (lowestTime - .05)*1000
    #snip off excess audio so they are all equal 
    extract = audio[startTime:endTime]
    extract.export("C://users//ericc//Desktop//MLP3//newTest//" + i, format="mp3")
```
