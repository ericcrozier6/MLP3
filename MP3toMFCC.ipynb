```python
import os
import librosa as lb
from scipy.fft import fft, fftfreq
import scipy
import numpy as np
```

```python
ls = []
counter = 0
#loop through files
for fileName in os.scandir(os.path.abspath("./project3/test")):
    #load mp3, trim to 29 seconds
    y, sr = lb.load(os.path.abspath(fileName), duration=29)
    #
    mfc = lb.feature.mfcc(y=y, sr=sr)
    x = np.mean(mfc[2:18], axis = 0)
    ls.append(x)
    #print(str(counter))
    #counter += 1
print(ls)
```

```python
#convert to np
a = np.vstack(ls)
#convert to npz
sparse_matrix = scipy.sparse.csc_matrix(a)
scipy.sparse.save_npz('test.npz', sparse_matrix)
print("done")
```
